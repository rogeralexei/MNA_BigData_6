```python
import pandas as pd
from tabulate import tabulate

games = pd.read_csv('./steam_games.csv')

print(tabulate(games.head(), headers='keys', tablefmt='grid'))
```

```python
game_review_counts = games['recent_reviews'].value_counts()
print(game_review_counts.describe())

game_review_counts = games['original_price'].value_counts()
print(game_review_counts.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

print(games.describe())

# Visualización de la distribución de reseñas por juego
sns.histplot(game_review_counts, bins=100, kde=True)
plt.title('Distribución de Reseñas por Juego')
plt.xlabel('Número de Reseñas')
plt.ylabel('Frecuencia')
plt.show()

# Visualización de precios de paquetes
sns.boxplot(x=games['original_price'])
plt.title('Distribución de Precios')
plt.xlabel('Precio por juego')
plt.show()
```
